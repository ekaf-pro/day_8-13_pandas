```python
# SETUP CELL - ‼️JANGAN DIUBAH‼️
# Jalankan cell ini terlebih dahulu untuk menghasilkan dataset tugas

import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Set random seed untuk reproducibility
np.random.seed(42)

print(">>> Menghasilkan dataset tugas...")

# 1. Generate Dataset Customers
def generate_customers(n=800):
    customer_data = {
        'customer_id': range(1, n+1),
        'nama_depan': np.random.choice(['Andi', 'Sari', 'Budi', 'Dewi', 'Rizki', 'Maya', 'Doni', 'Fitri'], n),
        'nama_belakang': np.random.choice(['Pratama', 'Wijaya', 'Santoso', 'Putri', 'Kusuma', 'Sari', 'Nugroho'], n),
        'umur': np.random.normal(35, 12, n).astype(int).clip(18, 65),
        'jenis_kelamin': np.random.choice(['L', 'P'], n),
        'email_domain': np.random.choice(['gmail.com', 'yahoo.com', 'outlook.com', 'perusahaan.com'], n, p=[0.4, 0.25, 0.25, 0.1]),
        'kota': np.random.choice(['Jakarta', 'Surabaya', 'Bandung', 'Medan', 'Semarang', 'Makassar'], n),
        'segmen_pelanggan': np.random.choice(['Premium', 'Standard', 'Basic'], n, p=[0.2, 0.5, 0.3]),
        'tanggal_daftar': pd.date_range('2023-01-01', '2024-12-31', periods=n)
    }
    df = pd.DataFrame(customer_data)
    df['email'] = df['nama_depan'].str.lower() + '.' + df['nama_belakang'].str.lower() + '@' + df['email_domain']
    df['nama_lengkap'] = df['nama_depan'] + ' ' + df['nama_belakang']
    return df[['customer_id', 'nama_lengkap', 'umur', 'jenis_kelamin', 'email', 'kota', 'segmen_pelanggan', 'tanggal_daftar']]

# 2. Generate Dataset Products
def generate_products(n=150):
    kategori = ['Elektronik', 'Pakaian', 'Buku', 'Olahraga', 'Rumah Tangga', 'Kecantikan']
    merek = ['MerekA', 'MerekB', 'MerekC', 'MerekD', 'MerekE']

    product_data = {
        'product_id': range(1, n+1),
        'nama_produk': [f'Produk_{i}' for i in range(1, n+1)],
        'kategori': np.random.choice(kategori, n),
        'merek': np.random.choice(merek, n),
        'harga': np.random.lognormal(3, 0.8, n).round(0),
    }

    df = pd.DataFrame(product_data)
    df['biaya'] = (df['harga'] * np.random.uniform(0.4, 0.7, len(df))).round(0)
    df['margin_keuntungan'] = ((df['harga'] - df['biaya']) / df['harga'] * 100).round(2)
    return df

# 3. Generate Dataset Orders
def generate_orders(customers_df, products_df, n=2000):
    order_data = {
        'order_id': range(1, n+1),
        'customer_id': np.random.choice(customers_df['customer_id'], n),
        'product_id': np.random.choice(products_df['product_id'], n),
        'jumlah': np.random.choice([1, 2, 3, 4, 5], n, p=[0.5, 0.25, 0.15, 0.07, 0.03]),
        'tanggal_order': pd.date_range('2024-01-01', '2024-12-31', periods=n),
        'status': np.random.choice(['Selesai', 'Pending', 'Dibatalkan'], n, p=[0.85, 0.1, 0.05])
    }
    return pd.DataFrame(order_data)

# 4. Generate Dataset Sessions
def generate_sessions(customers_df, n=3000):
    session_data = {
        'session_id': range(1, n+1),
        'customer_id': np.random.choice(customers_df['customer_id'], n),
        'waktu_mulai_sesi': pd.date_range('2024-01-01', '2024-12-31', periods=n),
        'halaman_dilihat': np.random.poisson(5, n) + 1,
        'durasi_sesi_menit': np.random.exponential(15, n).round(1),
        'tipe_device': np.random.choice(['Desktop', 'Mobile', 'Tablet'], n, p=[0.5, 0.4, 0.1]),
        'konversi': np.random.choice([True, False], n, p=[0.15, 0.85])
    }
    df = pd.DataFrame(session_data)
    df['bounce_rate'] = (df['halaman_dilihat'] == 1).astype(int)
    return df

# Generate semua dataset
customers_df = generate_customers(800)
products_df = generate_products(150)
orders_df = generate_orders(customers_df, products_df, 2000)
sessions_df = generate_sessions(customers_df, 3000)

print("Pembuatan dataset selesai!")
print(f"-- Customers: {len(customers_df)} baris")
print(f"-- Products: {len(products_df)} baris")
print(f"-- Orders: {len(orders_df)} baris")
print(f"-- Sessions: {len(sessions_df)} baris")
print("\nSiap memulai tugas Anda!")
```

```python
df['tanggal_daftar'] = pd.to_datetime(df['tanggal_daftar'])
today = pd.Timestamp.today()
df['tenure_hari'] = (today - df['tanggal_daftar']).dt.days
df
```

```python
df['tanggal_order'] = pd.to_datetime(df['tanggal_order'])
df = df.sort_values(['customer_id','tanggal_order'])
df['selisih_hari'] = (df.groupby('customer_id')['tanggal_order'].diff().dt.days)
avg_interval = (df.groupby('customer_id').agg(avg_hari_antar_order=('selisih_hari','mean')).reset_index())
df = pd.merge(df, avg_interval, on = 'customer_id', how='left')
df
```

```python
df['hari_order'] = df['tanggal_order'].dt.day_name()
freq = (df.groupby(['customer_id','hari_order']).size().reset_index(name='jumlah'))
idx = freq.groupby('customer_id')['jumlah'].idxmax()
fav_day = freq.loc[idx][['customer_id','hari_order']]
fav_day = fav_day.rename(columns={'hari_order':'hari_favorit'})
df = pd.merge(df,fav_day, on='customer_id', how='left')
df
```

```python
total_spend = (df.groupby('customer_id')['revenue'].sum().reset_index())
total_spend['tier'] = pd.qcut(total_spend['revenue'],q=3,labels=['Low','Medium','High'])
df = pd.merge(df,total_spend[['customer_id','tier']], on = 'customer_id', how ='left')
df
```
