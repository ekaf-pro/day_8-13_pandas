## 2. Group by, Join, dan Merge Data

### 2.1 GroupBy agregasi
Hitung **total `amount` per `city`**, urutkan dari terbesar.

```python
# Tulis di sini
(orders.groupby("city", as_index=False)["amount"].sum()
        .rename(columns={"amount": "total_sales"})
        .sort_values("total_sales", ascending=False)
)
```

### 2.2 Join (index-based) untuk lookup
Buat lookup **region per city**, lalu join ke `orders` (left join by index):
- Jakarta/Bandung/Surabaya → Jawa
- Medan → Sumatra
- Denpasar → Bali

```python
# Tulis di sini
city_lookup = pd.DataFrame({
    "city": ["Jakarta", "Bandung", "Surabaya", "Medan", "Denpasar"],
    "region": ["Jawa", "Jawa", "Jawa", "Sumatra", "Bali"]
}).set_index("city")

city_lookup
```

```python
orders.join(city_lookup, on="city")
```

### 2.3 Merge (column-based) dengan Products
Gabungkan `orders` dengan `products` untuk mendapatkan product_name. Tampilkan `order_id`, `product_name`, `product_category`, `amount`.

```python
orders.head(3)
```

```python
products.head(3)
```

```python
# Tulis di sini
orders_prd = pd.merge(orders,
         products[["product_id", "product_name", "product_category"]],
         on="product_id",
         how="left"
         )
orders_prd[["order_id", "product_name", "product_category", "amount"]]
```

### 2.4 Merge ke customers lalu cari top customer per city
Gabungkan `orders` ↔ `customers` pada `customer_id`, hitung total `amount` per (city, customer_name), lalu ambil 1 customer teratas per kota.

```python
# Tulis di sini
orders_cust = pd.merge(
    orders, customers[["customer_id", "customer_name", "city"]]
    .rename(columns={"city": "customer_city"}),
    on="customer_id", how="left"
)

city_customer_sales = orders_cust.groupby(["city", "customer_name"], as_index=False)["amount"].sum().sort_values(["city", "amount"], ascending=[True, False])

top_per_city = city_customer_sales.groupby("city", as_index=False).head(1)
top_per_city
```

## 3. Datetime

### 3.1 Konversi tanggal & komponen
Ubah `order_date` jadi datetime; buat kolom `year`, `month`, dan `day_name`.

```python
# Tulis di sini
orders["order_date"] = pd.to_datetime(orders["order_date"])

orders["year"] = orders["order_date"].dt.year
orders["month"] = orders["order_date"].dt.month
orders["day_name"] = orders["order_date"].dt.day_name()

orders[["order_id", "order_date", "year", "month", "day_name"]]
```

### 3.2 Filter range tanggal
Ambil transaksi dari **2024-02-01 s/d 2024-02-02**.

```python
# Tulis di sini
mask = (orders["order_date"] >= "2024-02-01") & (orders["order_date"] <= "2024-02-02")
orders.loc[mask, ["order_id", "order_date", "city", "amount"]]
```

### 3.3 Monthly sales dengan resampling
Jadikan `order_date` sebagai index dan hitung total `amount` per bulan, sekaligus jumlah order per bulan.

```python
orders
```

```python
# Tulis di sini
orders_dt = orders.copy()
orders_dt["amount"] = orders_dt["qty"] * orders_dt ["price"]
orders_dt = orders_dt.set_index("order_date")

monthly = (orders_dt
           .resample("ME")    # ME --> month end; MS --> month start
            .agg(total_amonut=("amount", "sum"),
                 order_count=("order_id", "count"))
            .reset_index()
)
monthly
```

## 6. Feature Engineering

### 6.1 Label Encoding sederhana
Buat peta untuk `product_category`: `{"Food":0,"Beverage":1,"NonFood":2}` → kolom `cat_le`.

```python
# Tulis di sini
cat_map = {"Food": 0, "Beverage": 1, "NonFood": 2}
orders["cat_le"] = orders["product_category"].map(cat_map)
orders[["order_id", "product_category", "cat_le"]]
```

### 6.2 One-Hot Encoding `city` dan binning `price`
- OHE city → kolom `city_...`
- Binning `price` ke low/mid/high (≤15k, 15k–30k, >30k) → price_bucket

```python
# Tulis di sini
city_ohe = pd.get_dummies(orders["city"], prefix="city")
orders_fe = pd.concat([orders, city_ohe], axis=1)

orders_fe["price_bucket"] = pd.cut(
    orders_fe["price"],
    bins=[0, 15_000, 30_000, float("inf")],
    labels=["low", "mid", "high"],
    right=True
)
orders_fe[["order_id", "city", "price", "price_bucket"] + list(city_ohe.columns)]
```

### 6.3 Target mean encoding sederhana untuk kategori
Hitung rata-rata `amount` per `product_category`, lalu map kembali ke setiap baris sebagai kolom `cat_mean_amount`.

```python
# Tulis di sini
cat_mean_amount = orders.groupby("product_category")["amount"].mean().to_dict()
orders["cat_mean_amount"] = orders["product_category"].map(cat_mean_amount)
orders
```
