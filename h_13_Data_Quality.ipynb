```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("transaksi.csv")

# --- 1) Heatmap Missing Value per Kolom vs order_id ---
cols = ["order_id", "customer_id", "product_id", "jumlah", "harga", "alamat_pengiriman"]
m = df[cols].isna().astype(int)  # 1 = missing, 0 = filled

plt.figure()
plt.imshow(m.T, aspect="auto", interpolation="nearest")
plt.title("Missing Value Heatmap")
plt.yticks(range(len(cols)), cols)
plt.xticks(range(len(df)), df["order_id"], rotation=90)
plt.xlabel("order_id")
plt.tight_layout()
plt.show()

# --- 2) Histogram Harga (deteksi outlier) ---
# Pastikan kolom harga numerik; non-numerik (mis. 'abc') jadi NaN
df["harga_num"] = pd.to_numeric(df["harga"], errors="coerce")

plt.figure()
plt.hist(df["harga_num"].dropna(), bins=30)
plt.title("Distribusi Harga Produk")
plt.xlabel("Harga (Rp)")
plt.ylabel("Frekuensi")
plt.tight_layout()
plt.show()

# Opsional: deteksi outlier sederhana (di atas p99)
p99 = df["harga_num"].quantile(0.99)
outliers = df.loc[df["harga_num"] > p99, ["order_id", "product_id", "harga_num"]]
print("Suspek outlier (harga > p99):")
print(outliers.to_string(index=False))

# --- 3) Distribusi Jumlah Order per Customer ---
order_per_cust = df.groupby("customer_id")["order_id"].nunique()

plt.figure()
plt.hist(order_per_cust, bins=20)
plt.title("Distribusi Jumlah Order per Customer")
plt.xlabel("Jumlah Order per Customer")
plt.ylabel("Jumlah Customer")
plt.tight_layout()
plt.show()

# Opsional: top customer (potensi reseller/fraud)
top_cust = order_per_cust.sort_values(ascending=False).head(10)
print("\nTop 10 customer berdasarkan jumlah order:")
print(top_cust)
```
